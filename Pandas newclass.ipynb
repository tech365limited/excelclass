```python
# adding new row
new_person = pd.Series({"No.":11.0,"Name": "Mark Zukerberg",  
                          "Net worth (USD)": "$54.7 billion", "Age": 25,
                          "Nationality": "United States", 
                          "Primary source(s) of wealth":  "Facebook",
                            "Year of Birth": 2023 - 35, 
                          "Place of birth": "Lagos"})
```

```python
df = df.append(new_person, ignore_index=True)
```

```python
#filtering data
#bringing out data of those in united states
df[(df["Nationality"] == "United States") & (df["Age"] > 70)]
```

```python
# renaming columns
df.rename(columns={"Nationality":"Country"})
```

```python
# replacing items in columns
df["Nationality"] = df["Nationality"].replace("United States", "USA")
df
```

```python
# applying custom function  
def clean_data(x):
    return float(x.replace("$", "").replace("billion", ""))
```

```python
clean_data("$54.7 billion")
```

```python
df["Networth"] =  df["Net worth (USD)"].apply(clean_data)
df
```
