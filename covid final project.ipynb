```python
df = df.set_index('date')
df
```

```python
# creating pivot table
flourish = pd.pivot_table(df,values="total_cases", index=["location"], columns="date").reset_index()
flourish = flourish.fillna(0)
flourish
```

```python
flags = pd.read_csv("https://raw.githubusercontent.com/rajeevratan84/covid/master/flags_world.csv")
flags
```

```python
# merge data with flag
florish = pd.merge(flourish, flags, left_on="location", right_on="country")
florish.head()
```

```python
del florish["country"]
images = florish["Image URL"]
florish.drop(labels=["Image URL"], axis=1, inplace=True)
florish.insert(1,"Flags", images)
florish
```

```python
florish.to_csv("covid_final_project.csv")
```
