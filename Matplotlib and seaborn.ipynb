```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
a = np.arange(1,13)
a
```

```python
b = 2 * a
b
```

```python
# labelling and saving your graph
plt.plot(a,b)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Graph Title")
plt.savefig("newgraph.jpg")
```

```python
# increasing the size of your graph
plt.figure(figsize=(12,6))
plt.plot(a,b)
```

```python
# styling your graph
plt.figure(figsize=(4,3))
plt.plot(a,a**2, label="x squared", color="red", lw=4)
plt.plot(a,a**3, label="x cube", marker='x', ms=10)
plt.legend(loc=2)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/waleolajumoke/data/main/dm_office_sales.csv")
df
```

```python
# scatterplot - check relationship between 2 or more values
sns.scatterplot(data=df, x="salary", y="sales", hue="division")
```

```python
# distribution plot
sns.displot(df, x="salary")
```

```python
sns.histplot(df,x="salary")
```

```python
# categorical plot
plt.figure(figsize=(12,5))
sns.countplot(data=df, x="division")
ax = plt.gca()
for i in ax.patches:
    ax.annotate(f"{i.get_height()}", (i.get_x() + i.get_width() / 2., i.get_height()))
```

```python
# categorical plot
plt.figure(figsize=(12,5))
sns.countplot(data=df, x="division", hue="level of education", palette="Blues")
```

```python
plt.figure(figsize=(12,5))
sns.barplot(data=df, x="level of education", y="salary")
```
